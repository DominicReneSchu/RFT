# Demo: Batch Study for Schu Resonance Field Theory

This notebook shows how to perform systematic parameter scans ("batch studies") using `schu_resonanzfeld.py` and save/analyse the results.

> **Prerequisites:**
> - schu_resonanzfeld.py on the Python path (e.g. in the same directory)
> - numpy, pandas, matplotlib installed


```python
import numpy as np
import pandas as pd
from schu_resonanzfeld import berechne_resonanzenergie, berechne_resonanzentropie
```

## 1. Define parameter ranges for the batch study


```python
# Finer or coarser spacing possible, depending on requirements
A_values = np.linspace(0.1, 5.0, 20)    # 20 amplitude values
T_values = np.linspace(0.1, 5.0, 20)    # 20 time constant values
```

## 2. Compute grid and collect results


```python
E_res, T_grid, A_grid = berechne_resonanzenergie(A_values, T_values)
S_grid = berechne_resonanzentropie(E_res)
```

## 3. Save results as DataFrame (e.g. for later analysis or seaborn plot)


```python
# The grids are reshaped into long ("melted") tabular form
result_df = pd.DataFrame({
    'A': A_grid.flatten(),
    'T': T_grid.flatten(),
    'E_res': E_res.flatten(),
    'S': S_grid.flatten(),
})
result_df.head()
```

## 4. Export as CSV (optional)


```python
result_df.to_csv('batch_resonanzfeld_results.csv', index=False)
print("Results saved as batch_resonanzfeld_results.csv.")
```

## 5. Example visualisation with seaborn (optional, if installed)


```python
try:
    import seaborn as sns
    import matplotlib.pyplot as plt
    plt.figure(figsize=(8,6))
    pivot = result_df.pivot_table(index='A', columns='T', values='E_res')
    sns.heatmap(pivot, cmap='inferno')
    plt.title('Resonance energy $E_{res}$ as a function of $A$ and $T$')
    plt.xlabel('Time constant $T$')
    plt.ylabel('Amplitude $A$')
    plt.show()
except ImportError:
    print("seaborn not installed — heatmap skipped.")
```
